```python
import numpy as np
import ipyvolume as ipv
import ipywidgets as widgets
```

```python
def s2c(Theta, Phi):
    return np.array([np.sin(Theta)*np.cos(Phi), np.sin(Theta)*np.sin(Phi), np.cos(Theta)])

def Lerp(a, b, t):
    return a + (b - a)*t

def Saturate(u):
    return np.minimum(1.0, np.maximum(0.0, u))

def VectorDot(a, b):
    return a[0]*b[0] + a[1]*b[1] + a[2]*b[2]

def Normalize(vec):
    l = np.sqrt(VectorDot(vec, vec))
    return np.array([vec[0]/l, vec[1]/l, vec[2]/l])
```

```python
# GGX - the final BRDF is multiplied by the cosine from the rendering equation - we refer to that as CBRDF in the notes!*
def SchlickFresnel(u):
    return np.power(Saturate(1 - u), 5.0)

def GTR2(NdotH, a):
    a2 = a*a
    denom = 1 + (a2 - 1) * NdotH * NdotH
    return a2/np.pi/(denom*denom)

def SmithG(NdotV, alphaG):
    return 2.0/(1 + np.sqrt(1 + alphaG*alphaG*(1 - NdotV*NdotV)/(NdotV*NdotV)))

def H(ThetaLight, PhiLight, ThetaView):
    L = s2c(ThetaLight, PhiLight)
    V = s2c(ThetaView, 0.0)
    for i in np.arange(3):
        L[i] += V[i]
    return Normalize(L)

def NdotH(ThetaLight, PhiLight, ThetaView):
    return H(ThetaLight, PhiLight, ThetaView)[2]

def LdotH(ThetaLight, PhiLight, ThetaView):
    Wh = H(ThetaLight, PhiLight, ThetaView)
    Wi = s2c(ThetaLight, PhiLight)
    return Wh[0]*Wi[0] + Wh[1]*Wi[1] + Wh[2]*Wi[2]

def RoughnessRemap(roughness):
    return (roughness*0.5 + 0.5)**2

def BRDF(ThetaLight, PhiLight, ThetaView, roughness, C0spec):
    # TODO: roughness is mapping differently for G and D term in mathemetica implemention, is that correct? Will review it later
    alphaG = RoughnessRemap(roughness)
    alpha = roughness*roughness

    G = SmithG(np.cos(ThetaLight), alphaG) * SmithG(np.cos(ThetaView), alphaG)
    F = Lerp(C0spec, 1, SchlickFresnel(LdotH(ThetaLight, PhiLight, ThetaView)))
    D = GTR2(NdotH(ThetaLight, PhiLight, ThetaView), alpha)
    denom = 4 * np.cos(ThetaLight)* np.cos(ThetaView)
    return F * D * G * np.maximum(np.cos(ThetaLight), 0) / denom
```

```python
def SphericalPlot3D(func, color=[1., 0, 0, 1.]):
    numsamples = 128
    
    theta = np.linspace(0, np.pi, num=numsamples)
    phi = np.linspace(0, 2*np.pi, num=numsamples*2)
    theta, phi = np.meshgrid(theta, phi)

    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)

    vals = func(theta, phi)
    if color[3] < 1.0:
        obj = ipv.plot_mesh(x*vals, z*vals, y*vals, wireframe=False, color=color)
        obj.material.transparent = True
        obj.material.side = "FrontSide"
    else:
        obj = ipv.plot_mesh(x*vals, z*vals, y*vals, wireframe=False, color=color)
```

```python
# our approximate model
def ApproxModelN(i, numLobes):
    return (i + 1) * 2 - (numLobes + 1)

def ApproxModelLobeTheta(i, numLobes, Theta, spread):
    return Theta + ((2 * i) / (numLobes - 1) - 1) * spread

def ApproxModelD(i, numLobes, Theta, spread):
    return ApproxModelN(i, numLobes) / 2 * (ApproxModelLobeTheta(i, numLobes, Theta, spread) - Theta) / spread

def ApproxModelLobeWeight(i, numLobes, Theta, spread):
    return np.exp(-(spread/3) * ApproxModelD(i, numLobes, Theta, spread)**2)

def ApproxModelLobe(i, numLobes, Theta, baseAmplitude, Lambda, spread,  ThetaLight, PhiLight): 
    return baseAmplitude * ApproxModelLobeWeight(i, numLobes, Theta, spread) * np.exp(-Lambda * (1 - VectorDot(s2c(ApproxModelLobeTheta(i, numLobes, Theta, spread), np.pi), s2c(ThetaLight, PhiLight))))

def ApproxModel(numLobes, Theta, baseAmplitude, Lambda, spread, ThetaLight, PhiLight):
    v = 0
    for i in np.arange(0, numLobes):
        v += ApproxModelLobe(i, numLobes,  Theta, baseAmplitude, Lambda, spread,  ThetaLight, PhiLight)
    return v
```

```python
def ThetaToFitIndex(dataSource, Theta):
     return np.floor((1 - np.cos(Theta)) * (dataSource.shape[1] - 1) + 0.5).astype(int)

def RoughnessToFitIndex(dataSource, roughness):
    return np.floor(roughness * (dataSource.shape[0] - 1) + 0.5).astype(int)

def ApproxModelFitGetData(dataSource, ThetaView, roughness):
    return dataSource[RoughnessToFitIndex(dataSource, roughness)][ThetaToFitIndex(dataSource, ThetaView)]

def ApproxModelFitGetTheta(dataSource, ThetaView, roughness):
    return ApproxModelFitGetData(dataSource, ThetaView, roughness)[0]

def ApproxModelFitGetAmplitude(dataSource, ThetaView, roughness): 
    return ApproxModelFitGetData(dataSource, ThetaView, roughness)[1]

def ApproxModelFitGetLambda(dataSource, ThetaView, roughness):
    return ApproxModelFitGetData(dataSource, ThetaView, roughness)[2]

def ApproxModelFitGetSpread(dataSource, ThetaView, roughness):
    return np.maximum(0.00001, ApproxModelFitGetData(dataSource, ThetaView, roughness)[3])

def ApproxModelFitLobeTheta(dataSource, i, ThetaView, roughness): 
    return ApproxModelLobeTheta(i, 8, ApproxModelFitGetTheta(dataSource, ThetaView, roughness), ApproxModelFitGetSpread(dataSource, ThetaView, roughness))

def ApproxModelFitEqualEnergy(ThetaLight, Philight, ThetaView, roughness): 
    return ApproxModel(8,
                       ApproxModelFitGetTheta(ApproxModelFitDataEqualEnergy, ThetaView, roughness), 
                       ApproxModelFitGetAmplitude(ApproxModelFitDataEqualEnergy, ThetaView, roughness),
                       ApproxModelFitGetLambda(ApproxModelFitDataEqualEnergy, ThetaView, roughness),
                       ApproxModelFitGetSpread(ApproxModelFitDataEqualEnergy, ThetaView, roughness),
                       ThetaLight, Philight)

def ApproxModelFitFreeAmplitude(ThetaLight, Philight, ThetaView, roughness): 
     return ApproxModel(8, 
                        ApproxModelFitGetTheta(ApproxModelFitDataFreeAmplitude, ThetaView, roughness), 
                        ApproxModelFitGetAmplitude(ApproxModelFitDataFreeAmplitude, ThetaView, roughness),
                        ApproxModelFitGetLambda(ApproxModelFitDataFreeAmplitude, ThetaView, roughness),
                        ApproxModelFitGetSpread(ApproxModelFitDataFreeAmplitude, ThetaView, roughness),
                        ThetaLight, Philight)
```

```python
def PlotTableData(data, transfo, component, sz):
    ipv.figure(lighting=False)
    xParams = np.linspace(0, 1, sz)
    zParams = np.linspace(0, 1, sz)
    xParams, zParams = np.meshgrid(xParams, zParams)

    vals = np.zeros_like(xParams)
    for i in np.arange(sz):
        for j in np.arange(sz):
            vals[i][j] = data[i][j][component]

    ipv.clear()
    ipv.plot_mesh(xParams, transfo(vals), zParams, wireframe=True)
#     ipv.xyzlim(0, 1)
    ipv.xlim(0, 1)
    ipv.zlim(0, 1)
#     ipv.ylim(0, 4)
    ipv.show()
```
