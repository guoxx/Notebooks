```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_func(x):
    return x

def gamma_func(x):
    return np.power(x, 2.2)

def test_func(x, scalar):
    return pow(x,3)*scalar

def reverse_simple_tonemappingGamma(x):
	return x / (1.035 - x) * 0.187

x = np.linspace(0, 1, 64)
y0 = linear_func(x)
y1 = gamma_func(x)

scalar_val = 4
y2 = test_func(x, scalar_val)

y3 = reverse_simple_tonemappingGamma(x)

plt.figure(figsize=(4,16))
plt.plot(x, y0, "o", ms=1.0, label="linear")
plt.plot(x, y1, "o", ms=1.0, label="gamma")
plt.plot(x, y2, "o", ms=1.0, label="test")
plt.plot(x, y3, "o", ms=1.0, label="revertiable tonemapping gamma")

plt.plot([0.5, 1], [0.5, 4])

plt.plot([0, 1], [0.5, 0.5])
plt.legend()
plt.xlim(0, 1)
plt.ylim(0, 4)

print(np.sum(y0<=0.25))
print(np.sum(y1<=0.25))
print(np.sum(y2<=0.5))
print(np.sum(y3<=0.5))
```

```python
def Tonemap_UE3(x):
    color = x / (x + 0.187) * 1.035
    return pow(pow(color, 2.2), 0.45)

def Tonemap_ACES(x):
    a = 2.51;
    b = 0.03;
    c = 2.43;
    d = 0.59;
    e = 0.14;
    ret = (x * (a * x + b)) / (x * (c * x + d) + e);
    return pow(ret, 0.45)

x = np.arange(0, 2.5, 0.01)
y = Tonemap_UE3(x)
y1 = Tonemap_ACES(x)
plt.figure(figsize=(8,4))
plt.plot(x, y, label="UE3")
plt.plot(x, y1, label="ACES")
plt.xlim(0, 2)
plt.ylim(0, 1)
plt.legend()
print(Tonemap_UE3(2.5))
```
